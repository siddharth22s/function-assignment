**1. What is the difference between a function and a method in Python?**

- Key Differences:
1. Functions are independent, while methods belong to a class.

2. Methods require an instance (self) to operate on class attributes.

3. Functions can be used anywhere, while methods are tied to objects.

>> Function >>

1. Defined outside of a class.

2. Can be called independently using its name.

3. Does not require an object to be invoked.


example----

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("siddharth"))
```

>> Method >>

1. Defined inside a class.

2. Called on an object using dot notation.

3. Can access and modify class attributes.


example ---

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

p = Person("siddharth")
print(p.greet())
```

**6. Explain the concept of generators in Python and how they are defined.**

- Generators are special functions that return an iterator and allow lazy evaluation, meaning they generate values on demand instead of storing them in memory.
A generator function is defined like a normal function but uses yield instead of return:

```python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()
print(next(gen))
print(next(gen))
print(next(gen))
```

**7. What are the advantages of using generators over regular functions?**

- Generators provide several benefits compared to regular functions, especially when dealing with large datasets or iterative computations.

1. Memory Efficient > Uses less memory by generating values one at a time.
2. Lazy Evaluation > Computes values only when needed.  
3. Faster Performance > Avoids unnecessary computations for large data.
4. Simpler Iteration > No need for manual state tracking.


```python
def square_numbers(n):
    return [x*x for x in range(n)]
print(square_numbers(5))
def square_numbers_gen(n):
    for x in range(n):
        yield x*x
gen = square_numbers_gen(5)
print(next(gen))
print(next(gen))
print(next(gen))
```

**9. Explain the purpose and usage of the `map()` function in Python.**

- The map() function applies a given function to each item in an iterable (like a list or tuple) and returns an iterator with the transformed values.

- Syntax > map(function, iterable)


```python
# coverting str into integer
numbers = ['1', '2', '3', '4']
result = map(int, numbers)
print(list(result))
```

```python
# Using map() with Lambda
nums = [1, 2, 3, 4]
doubled = list(map(lambda x: x * 2, nums))
print(doubled)
```

```python
#Using map() with Multiple Iterables
a = [1, 2, 3]
b = [4, 5, 6]
summed = list(map(lambda x, y: x + y, a, b))
print(summed)  # Output: [5, 7, 9]
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

- map(), reduce(), and filter()â€”are built-in Python functions used for processing iterables efficiently. Here's how they differ:
map(function, iterable)
1. map(function, iterable)
Applies a given function to every item in an iterable (e.g., list, tuple) and returns a new iterator with transformed values.

```python
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, nums))
print(squared)  # Output: [1, 4, 9, 16]
```

2. reduce(function, iterable) (from functools module)

Applies a rolling computation to sequential pairs in an iterable, reducing it to a single value.

```python
from functools import reduce
nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)
```

4. filter(function, iterable)

Filters elements from an iterable based on whether they satisfy a given condition.

```python
nums = [1, 2, 3, 4, 5, 6]
even_nums = list(filter(lambda x: x % 2 == 0, nums))
print(even_nums)
```

**4. Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")
```

**5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        else:
            fib_num = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return fib_num


fib_sequence = FibonacciIterator(10)
print(list(fib_sequence))
```

**6. Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

gen = powers_of_two(5)
print(list(gen))
```

**7. Implement a generator function that reads a file line by line and yields each line as a string.**

```python
def read_file_lines(file_path):
  with open(file_path, 'r') as file:
    for line in file:
      yield line

with open("my_file.txt", "w") as f:
  f.write("This is the first line.\n")
  f.write("This is the second line.\n")
  f.write("And this is the third line.")

for line in read_file_lines("my_file.txt"):
  print(line, end='')
```

**9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temperatures = [0, 20, 37, 100]
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))
print(fahrenheit_temperatures)
```

**10. Create a Python program that uses `filter()` to remove all the vowels from a given string.**

```python
def is_not_vowel(char):
    return char.lower() not in "aeiou"
input_string = "Hello, Siddharth! Python is fun."
filtered_string = "".join(filter(is_not_vowel, input_string))
print(filtered_string)
```
